```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
training_df = pd.read_csv('/Volumes/DataDrive/training/logs/logs/train-1.csv')
```

```python
training_df.head(n=5)
```

```python
acc = training_df['acc'].values
val_acc = training_df['val_acc'].values
loss = training_df['loss'].values
val_loss = training_df['val_loss'].values

# Plot accuracy
plt.plot(acc)
plt.plot(val_acc)
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper left')
plt.show()

# Plot loss
plt.plot(loss)
plt.plot(val_loss)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='upper right')
plt.show()
```

```python
f2_score = training_df['f2_score'].values

# Plot F2 Score
plt.plot(f2_score)
plt.title('Wartość F2')
plt.ylabel('Wartość')
plt.xlabel('Epoka')
plt.legend(['Trening'], loc='upper left')
plt.show()
```
